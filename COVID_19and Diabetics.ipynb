```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

class DiabetesDataAnalyzer:
    def __init__(self, state_name):
        self.state_name = state_name
        self.filename = f"DataSet/DiabetesAtlas_{state_name}.csv"
        self.dataframe = self.process_state_data()

    def process_state_data(self):
        try:
            df = pd.read_csv(self.filename)
            print(f"First few lines of {self.filename}:")
            print(df.head())
            df = df[['Year', '18-44 - Percentage', '45-64 - Percentage', '65-74 - Percentage', '75+ - Percentage']]
            df['Year'] = pd.to_datetime(df['Year'], format='%Y', errors='coerce') 
            df = df.dropna(subset=['Year'])  
            df.set_index('Year', inplace=True)
            return df
        except pd.errors.ParserError as e:
            print(f"Error reading file {self.filename}: {e}")
            return None

    def plot_percentage_change(self):
        if self.dataframe is None:
            return
        fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
        fig.suptitle(f"{self.state_name} - Diabetic Percentage Change Before and After COVID-19")

        for i, col in enumerate(self.dataframe.columns):
            ax = axes[i // 2, i % 2]
            df_before = self.dataframe.loc['2018':'2019']  
            df_after = self.dataframe.loc['2019':'2020']   
            ax.plot(df_before.index, df_before[col], label='Before COVID-19')
            ax.plot(df_after.index, df_after[col], label='After COVID-19')
            ax.set_xlabel('Year')
            ax.set_ylabel('Percentage')
            ax.set_title(col)
            ax.legend()

        plt.tight_layout()
        plt.show()

    def correlation_analysis(self):
        if self.dataframe is None:
            return
        df_before = self.dataframe.loc['2018':'2019']  
        df_after = self.dataframe.loc['2019':'2020']   
        df_before = df_before.dropna()
        df_after = df_after.dropna()
        if len(df_before) < 2 or len(df_after) < 2:
            print(f"Not enough data for correlation analysis for {self.state_name}.")
            return
        
        correlations = {}
        for col in df_before.columns:
            correlation = np.corrcoef(df_before[col], df_after[col])[0, 1]
            correlations[col] = correlation
        print(f"Pearson correlation analysis for {self.state_name}:")
        print(correlations)
        print()


state_names = ['Alabama', 'Colorado', 'Connecticut', 'Florida', 'Louisiana', 'Michigan', 'Mississippi', 'Montana', 'Missouri', 'Ohio', 'Pennsylvania', 'South Carolina']


for state_name in state_names:
    analyzer = DiabetesDataAnalyzer(state_name)
    analyzer.plot_percentage_change()
    analyzer.correlation_analysis()
```
